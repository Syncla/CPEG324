```python
def bin_add(*args): return bin(sum(int(x, 2) for x in args))[2:]

binary = []
for n in range(256):
    binary.append(bin(n))
```

```python
##create testbench cases for 8 bit adder
outstring = ""
file = open("outputcases.txt","w")
for m in binary:
    for n in binary:
        result = bin_add(m,n)
        a = m[2:]
        b = n[2:]
        while (len(a)<8):
            a = "0"+a
        while (len(b)<8):
            b = "0"+b            
        while (len(result)<8):
            result = "0"+result
        while (len(result)>8):
            result = result[1:]
        file.write("(\""+str(a)+"\","+"\""+str(b)+"\","+"\""+result+"\"),\n")
file.close()
```

```python
##testbench cases for zerocheck
for m in binary:
    inpt = m[2:]
    while (len(inpt)<8):
        inpt = "0"+inpt
    if (inpt=="00000000"):
        print("(\""+inpt+"\","+"\"1\"),")
    elif (inpt !="00000000"):
        print("(\""+inpt+"\","+"\'0\'),")
```
